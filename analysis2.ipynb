```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# loading the data
df =  pd.read_csv(r'/content/drive/MyDrive/united2/RAw/sentiment.csv')
```

```python
df.head(3)
```

```python

# Create a new column 'Sentiment_Category' based on the 'average_sentiment' values
df['Sentiment_Category'] = 'Positive'
df.loc[df['average_sentiment'] < 0, 'Sentiment_Category'] = 'Negative'
df.loc[(df['average_sentiment'] >= 0) & (df['average_sentiment'] <= 0.5), 'Sentiment_Category'] = 'Neutral'

# Count the occurrences of each sentiment category
sentiment_counts = df['Sentiment_Category'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Customer Sentiment Distribution')
plt.axis('equal')
plt.show()
```
